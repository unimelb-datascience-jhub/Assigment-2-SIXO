To simply the analysis, only sub set part of columns from weather dataset. As the columns regarding 9am and 3pm are specific to two timeslots of a day, the analysis is actually focusing on the daily records. Therefore, those columns are excluded for now.

```python
weather_subset = weather[['Date', 'Minimum temperature (°C)', 'Maximum temperature (°C)',
       'Rainfall (mm)', 'Evaporation (mm)', 'Sunshine (hours)',
      #  'Direction of maximum wind gust ',
       'Speed of maximum wind gust (km/h)',
       'Time of maximum wind gust',
       '9am Temperature (°C)',
       '9am relative humidity (%)', '9am cloud amount (oktas)',
    #    '9am wind direction', 
      '9am wind speed (km/h)', '9am MSL pressure (hPa)',
       '3pm Temperature (°C)', '3pm relative humidity (%)',
       '3pm cloud amount (oktas)',
      # '3pm wind direction',
       '3pm wind speed (km/h)', '3pm MSL pressure (hPa)'
    ]].copy()
```

```python
weather_subset.head()
```

```python
weather_subset['Date'] = pd.to_datetime(weather_subset['Date'], format='%d/%m/%Y')

weather_subset.dtypes
```

```python
weather_subset['Time of maximum wind gust']

# Define a function to convert 'Time of maximum wind gust'strings to a number of during time(in seconds)
def to_seconds(time_str):
    try: # the column of 'Time of maximun wind gust' contains float and str. Thus using try and except statement to handle the exception of float
        minutes, seconds = time_str.split(':')
        result = int(minutes) * 60 + int(seconds)
    except AttributeError:
        result = 0
    return result
    

# Convert the 'time' column to minutes using the to_minutes function
weather_subset['Seconds of maximum wind gust'] = weather_subset['Time of maximum wind gust'].apply(lambda x: to_seconds(x))

# Print the resulting dataframe
# print(weather_subset.head())

weather_subset.columns
```

```python
if weather_subset['Time of maximum wind gust'].dtype == 'object' and any(isinstance(val, float) for val in weather_subset['Time of maximum wind gust']):
    print("The column contains both floats and strings.")
else:
    print("The column does not contain both floats and strings.")
```

```python
weather_subset_updated = weather_subset.drop('Time of maximum wind gust', axis=1)
weather_subset_updated.head()
weather_subset_updated.set_index('Date',inplace=True)
weather_subset_updated.head()
```

```python
mask = weather_subset_updated.isna().any(axis=1)
row_with_na = weather_subset_updated.loc[mask]
row_with_na
```

```python


# data cleansing. Assume the value of'Calm' as the columns of wind speed means no wind.Then replace it with zero
weather_subset_updated['3pm wind speed (km/h)'] = weather_subset_updated['3pm wind speed (km/h)'].replace('Calm', 0.0)
weather_subset_updated['9am wind speed (km/h)'] = weather_subset_updated['9am wind speed (km/h)'].replace('Calm', 0.0)
weather_subset_updated.dtypes

# weather_subset_updated.dtypes
```

```python
# weather_subset_updated.dtypes
weather_subset_updated['3pm wind speed (km/h)'] = weather_subset_updated['3pm wind speed (km/h)'].astype('float64')
weather_subset_updated['9am wind speed (km/h)'] = weather_subset_updated['9am wind speed (km/h)'].astype('float64')
```

```python
# loop over the rows and fill the missing value with the average of preceding and following rows
for row in range(1, len(weather_subset_updated)-1):
    if weather_subset_updated.iloc[row].isna().any():
        #if the current row contains NA values, loop to the preceding and following rows
        prev_row = row - 1
        while prev_row > 0 and weather_subset_updated.iloc[prev_row].isna().any():
            prev_row -= 1
        follow_row = row + 1
        while follow_row < len(weather_subset_updated)-1 and weather_subset_updated.iloc[follow_row].isna().any():
            follow_row += 1
        weather_subset_updated.iloc[row] = (weather_subset_updated.iloc[prev_row] + weather_subset_updated.iloc[follow_row])/2

print(weather_subset_updated)

```

```python
dummy_price_category = pd.get_dummies(merged['PRICECATEGORY'], prefix='price_cat',drop_first=True)
dummy_price_category
```

```python
merged = pd.concat([merged, dummy_price_category], axis=1)
merged.drop('PRICECATEGORY', axis=1, inplace=True)
merged.head()
```

```python
# continue cleaning the dataset to ensure all the columns are named properly(capitalized the first letter)
merged.columns = merged.columns.str.title()
merged.columns
```

```python
# continue cleaning the column names
merged.rename(columns={
                        'Price_Cat_High': 'High Price',\
                        'Price_Cat_Low': 'Low Price',\
                        'Price_Cat_Medium': 'Medium Price'}, inplace=True)
merged.rename(columns={'Totaldemand': 'Total Demand'}, inplace=True)

merged.columns
```

```python
merged.head()
```

```python
#convert column of 'Date' to numeric values
# merged['Year'] = merged['Date'].dt.year
merged['Month'] = merged['Date'].dt.month
merged['Day'] = merged['Date'].dt.day

# merged['Year'] = merged['Year'].astype('category').cat.codes
merged['Month'] = merged['Month'].astype('category').cat.codes
merged['Day'] = merged['Day'].astype('category').cat.codes

merged.head(48)
```

```python
# convert the 'Time' column to numeric values

merged['Time'] = pd.to_datetime(merged['Time'], format='%H:%M').dt.time

# Define a function to convert time to numerical values
def time_to_num(time):
    hours = time.hour
    minutes = time.minute
    return int((hours * 60 + minutes) / 30)

# Apply the function to the 'Time' column
merged['Time'] = merged['Time'].apply(time_to_num)

# Print the updated data
merged.head()

```
